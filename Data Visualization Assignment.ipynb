```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.decomposition import PCA 
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
x = iris.iloc[:,[0,1,2,3]].values
y = iris.loc[:,'species'].values
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=3)
x_1 = pca.fit_transform(x)
```

```python
x_1
```

```python
df = pd.DataFrame(x_1,columns=['Principal_Component1','Principal_Component2','Principal_Component3'])
df.head()
```

```python
df1 = pd.DataFrame(y,columns=['Target'])
df1.head()
```

```python
df = pd.concat([df,df1],axis=1)
df.head()
```

```python
target_color = {'versicolor':'r','setosa':'b','virginica':'g'}
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Principal Component1')
ax.set_ylabel('Principal Component2')
ax.set_zlabel('Principal Component3')
ax.set_title('3D Visualization of Iris Data')
l=[]

for i in range(len(df)):
    for j in target_color:
        if j==df.loc[i,'Target']:
            c = target_color[j]
            if j not in l:
                l.append(j)
                ax.scatter(df.loc[i,'Principal_Component1'],df.loc[i,'Principal_Component2'],df.loc[i,'Principal_Component3'],color=c,label=j)
            else:
                 ax.scatter(df.loc[i,'Principal_Component1'],df.loc[i,'Principal_Component2'],df.loc[i,'Principal_Component3'],color=c)            
ax.legend()
    

```
